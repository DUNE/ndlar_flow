```python
import h5py
import numpy as np
import h5flow
from collections import defaultdict
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (8,6)
plt.rcParams.update({'font.size': 18})
```

```python
def find_tracks_in_packet(packet_num, flow_out):
    print("packet at index", packet_num,':')
    print('     ',flow_out['charge/packets/data'][packet_num])
    
    track_contr = []
    total = 0.
    frac_indices = np.where(flow_out['/charge/packets/ref/mc_truth/packet_fraction/ref'][:,0] == packet_num)[0]
    for i in frac_indices:
        total += flow_out['mc_truth/packet_fraction/data'][i]
        track_index = flow_out['charge/packets/ref/mc_truth/tracks/ref'][i,1]
        contrib = flow_out['mc_truth/packet_fraction/data'][i]
        track = flow_out['mc_truth/tracks/data'][track_index]
        #print(f'trackID {track_index} contributes {contrib*100}%')
        track_contr.append([track,contrib])
    #print('total','{}%'.format(total*100))
    return track_contr
```

```python
def find_tracks_in_packet(packet_num, flow_out):
    print("packet at index", packet_num,':')
    print('     ',flow_out['charge/packets/data'][packet_num])
    
    track_contr = []
    total = 0.
    frac_indices = np.where(flow_out['/charge/packets/ref/mc_truth/packet_fraction/ref'][:,0] == packet_num)[0]
    for i in frac_indices:
        total += flow_out['mc_truth/packet_fraction/data'][i]
        track_index = flow_out['charge/packets/ref/mc_truth/tracks/ref'][i,1]
        contrib = flow_out['mc_truth/packet_fraction/data'][i]
        track = flow_out['mc_truth/tracks/data'][track_index]
        #print(f'trackID {track_index} contributes {contrib*100}%')
        track_contr.append([track,contrib])
    #print('total','{}%'.format(total*100))
    return track_contr

def find_tracks_in_calib_final_hits(hit_num, flow_out):
    print("calib_final_hit at index", hit_num,':')
    print('     ',flow_out['charge/calib_final_hits/data'][hit_num])
    
    track_contr = []
    total = 0.

    prompt_region = flow_out['/charge/calib_final_hits/ref/charge/calib_prompt_hits/ref_region'][hit_num]
    #print('prompt_region:',prompt_region)
    prompt_charge = [flow_out['charge/calib_prompt_hits/data'][prompt_hit_id]['Q'] for prompt_hit_id in prompt_region]
    packet_numbers = [np.where(flow_out['/charge/calib_prompt_hits/ref/charge/packets/ref'][:,0] == prompt_hit)[0][0] for prompt_hit in range(prompt_region[0],prompt_region[1])]
    #print('prompt_charge', prompt_charge)
    #print('packet_numbers:',packet_numbers)
    norm = 0.
    for it, packet in enumerate(packet_numbers):
        pack_track_contr = find_tracks_in_packet(packet,flow_out)
        for t in pack_track_contr:
            track_contr.append([prompt_charge[it],t[1],t[0]]) # prompt_hit_charge, track_fraction, track
            norm+=track_contr[-1][0]*track_contr[-1][1]
    #print(track_contr)
    #TODO: get the charge of the associated prompt hits and do the weighting here
    track_contributions = []
    for track in track_contr:
        contr = track[0]*track[1]/norm
        track_contributions.append([contr,track[2]])
    # merge unique track contributinos
    track_dict = defaultdict(lambda:0)
    for track in track_contributions:
        track_dict[tuple(track[1])] += track[0]
    return track_dict

final_hit_tracks = find_tracks_in_calib_final_hits(10,flow_out)

print(final_hit_tracks)
```
